# Matplotlib example workflow


```python
# 0. import matplotlib and get it ready for plotting in Juptyter
%matplotlib inline 
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize= (10, 10))
        # figsize is just width & height
# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title = "Simple plot",
       xlabel= "x-axis",
       ylabel="y-axis")
#5 Save & show (you save the whole figure)
fig.savefig("..")
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x
```

```python
# Plot the data and create a line plot
fig, ax = plt.subplots()
ax.plot(x , x**2);
```

```python
fig, ax = plt.subplots()
ax.plot(x , np.exp(x));
```

```python
# Use same data to make a scatter
fig , ax = plt.subplots()
ax.scatter(x , np.exp(x));
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x , np.sin(x));
```

```python
fig, ax = plt.subplots()
ax.plot(x, np.sin(x));
```

```python
fig, ax = plt.subplots()
ax.bar(x , np.sin(x));
```

```python
# Make a plot from dictionary
nut_butter_prices ={"Amond butter": 10,
                    "Peanut butter": 8,
                    "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values() ) ;
ax.set(title = "Dan's Nut Butter Store",
      ylabel= "Price($)");
```

```python
fig, ax = plt.subplots()
building_prices= {"a": 124, "b": 345, "c": 32, "d": 231, "e": 234, "f": 345}
ax.scatter(building_prices.keys(), building_prices.values())
ax.set(title="Building prices",
      ylabel="Prices($)",
      xlabel=("Building Names"));
```

```python
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
```

### Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols= 2,
                                            figsize=(10, 5))

#Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
fig , ((ax1, ax2,ax3) ,(ax4, ax5, ax6), (ax7, ax8, ax9)) =plt.subplots(nrows=3,
                                                                      ncols=3,
                                                                      figsize= (10,5))
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
ax5.plot(x, x **2);
ax6.scatter(x, np.cos(x));
ax7.hist((x -3));
ax8.hist(np.random.random(100));
ax9.plot(np.sin(x));
```

```python
# Subplot Option 2

fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize = (10, 5))
# Plot to each different axis
ax[0,0].plot(x, x/2);
ax[0,1].plot(np.random.random(100));
ax[1,0].hist(np.random.randn(1000));
ax[1,1].scatter(x, x % 10);
```

```python
car_sales["Price"] = car_sales['Price'].str.replace('[\$\,\.]', '')
car_sales
```

```python
type(car_sales['Price'])
```

```python
type(car_sales['Price'][0])
```

```python
# Remove last two zeros
car_sales['Price'] = car_sales['Price'].str[:-2]
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2023", periods= len(car_sales) )
car_sales
```

```python
car_sales['Total Sales'] = car_sales['Price'].astype(int).cumsum()
car_sales
```

```python
type(car_sales['Price'] [0])
```

```python
fig, ax = plt.subplots(figsize = (20, 10))
ax.plot(car_sales['Sale Date'], car_sales['Total Sales']);
ax.set(title= "Car Sales",
      xlabel="Dates",
      ylabel="Total Profit $");
```
